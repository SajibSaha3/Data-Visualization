# subplot

```python
x = [3,4,5,9,6]
y = [6,8,7,9,2]
z= [4,8,2,3,8]

plt.subplot(1,3,1)
plt.plot(x,y, color = 'g')
plt.xlabel("lower value")
plt.ylabel('count values')
plt.title('The 1st subplot')
plt.legend("a")


plt.subplot(1,3,2)
plt.plot(x,z, color = 'r')
plt.xlabel("upper value")
plt.ylabel('count values')
plt.title('The 2nd subplot')
plt.legend("2nd_plot")

plt.subplot(1,3,3)
plt.plot(z,y, color = 'b')
plt.xlabel("Middle value")
plt.ylabel('count values')
plt.title('The 3rd subplot')
plt.legend("3nd_plot")
plt.show()
```

```python
import numpy as np
a = np.arange(0,10,1)
b = 1*a + 10
c = 2*a + 10
d = 3*a + 10
```

```python
b
```

```python
plt.subplot(1,3,1)
plt.plot(b,c, linewidth = 2.0, marker = 'o', color ='r', alpha= .8)
plt.title('Graph 1')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line1'], )
plt.grid()

plt.subplot(1,3,2)
plt.plot(d,c, linewidth = 2.0, marker = '*', color ='b', alpha= .8)
plt.title('Graph 2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line2'], )
plt.grid()

plt.subplot(1,3,3)
plt.plot(b,d, linewidth = 2.0, marker = '.', color ='black', alpha= .7)
plt.title('Graph 3')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line3'], )
plt.grid()
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(b,c, linewidth = 2.0, marker = 'o', color ='r', alpha= .8)
plt.title('Graph 1')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line1'], )
plt.grid()

plt.subplot(3,1,2)
plt.plot(d,c, linewidth = 2.0, marker = '*', color ='b', alpha= .8)
plt.title('Graph 2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line2'], )
plt.grid()

plt.subplot(3,1,3)
plt.plot(b,d, linewidth = 2.0, marker = '.', color ='black', alpha= .7)
plt.title('Graph 3')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['line3'], loc = 'best') # loc best refers the optimal position to fit 
plt.grid()
plt.show()
```

# pie plot

```python
x = ['ami','tumi','she']
y = [10,5,18]

plt.scatter(x,y, color = 'r')
plt.scatter(x,y, c= 'b', s = 200, edgecolors= 'black', marker= '*', alpha= 1)
plt.title('plot value')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
label = ['Dog', 'cat', 'wolf', 'Lion']
size = [20,50,25,32]

plt.pie(size, labels = label, autopct = '%1.2f%%', shadow= True, startangle = 90, explode = (.3,0,0,0))
plt.title("pie Chart Demo")
plt.show()
```
