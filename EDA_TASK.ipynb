```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

```python
df['species'].value_counts()
```

# 2D plot

```python
#scatter plot 
df.plot(kind = 'scatter', x = 'sepal_length', y = 'petal_width')
plt.show()
```

```python
# 2-D Scatter plot with color-coding for each flower type/class.
sns.set_style('whitegrid');  
sns.FacetGrid(df, hue="species", height=4) \
    .map(plt.scatter, "sepal_length", "petal_width") \
    .add_legend();
plt.show()
```

# 3D plot 

```python
import plotly.express as px 
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',color='species')
fig.show()
```

# pair plot 

```python
sns.set_style("whitegrid");
sns.pairplot(df, hue="species", height=3);
plt.show()
```

```python
df.describe()
```

# box plot

```python
sns.boxplot(x='species',y='petal_length', data=df)
plt.show()
```

```python
df.boxplot(column="petal_width", by = "species")
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length", data=df, size=8)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', alpha=0.7)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x= 'sepal_length' , y='sepal_width', alpha=0.8, hue='species')
plt.title('sepal Length vs sepal Width')
plt.xlabel('sepal_length')
plt.ylabel('sepal Width')
plt.grid(True)
plt.show()
```

```python
species_counts = df['species'].value_counts()
plt.bar(species_counts.index, species_counts.values, color=['red', 'green',"black"])
plt.title('species each categories')
plt.xlabel('types')
plt.ylabel('Count')

plt.show()
```

```python
species_counts = df['species'].value_counts()

# Plotting the donut chart
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=["setosa", "verginica","versicolor"], autopct='%1.1f%%', startangle=90, colors=['gold', 'yellowgreen', 'lightcoral'])
# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis()  
plt.title('total distrubution for species')
plt.show()
```
